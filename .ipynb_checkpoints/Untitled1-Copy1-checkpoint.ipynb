#### - 라이브러리 선언 및 드라이버 설정

```python
# 라이브러리 선언
from selenium import webdriver
from bs4 import BeautifulSoup

# 드라이버 위치 설정
driver_loc = "C:/Program Files/Google/Chrome/Application/chromedriver.exe"

# 드라이버 옵션 설정
options = webdriver.ChromeOptions()
#헤드리스 설정 시 리눅스 같이 웹화면 없는경우에도 웹페이지 실행가능함
# options.add_argument('headless')
# options.add_argument("window-size=1920*1080")
# options.add_argument('disable-gpu')

options.add_argument('--headless')
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# 웹드라이버 정의
driver = webdriver.Chrome(driver_loc, options=options)
```

```python
def strToFloat(x):
    return float(x.replace(',',''))
```

### 4-1. EPS * PER 함수

```python
def eps_per(finSummary):
    if finSummary[3]['EPS(원)'] == '':
        pre_eps = float(finSummary[1]['EPS(원)'])
        now_eps = float(finSummary[2]['EPS(원)'])
        eps = (pre_eps + now_eps / 2)
        per = (float(finSummary[0]['PER(배)']) + float(finSummary[1]['PER(배)']) + float(finSummary[2]['PER(배)'])) / 3
        eps_per = eps * per
    else:
        pre_eps = float(finSummary[2]['EPS(원)'])
        now_eps = float(finSummary[3]['EPS(원)'])
        eps = (pre_eps + now_eps / 2)
        per = (float(finSummary[0]['PER(배)']) + float(finSummary[1]['PER(배)']) + float(finSummary[2]['PER(배)']) + float(finSummary[3]['PER(배)'])) / 4
        eps_per = eps * per
    return eps_per
```

```python
eps_per(finSummary)
```
